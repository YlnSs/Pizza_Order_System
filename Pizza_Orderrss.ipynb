```python
import csv
import datetime



#Greeting the client
def welcome_to_ordering_system():
  print("Welcome to pizza ordering system")
#For thanking to users
def see_you():
  print("thanks for using the system! Have a good one!")

def menu():
  with open("menu.txt",'r') as file:
    contents=file.read()
    print(contents)
#Almost everything is happening here we are taking orders and infos and saving them this is main function
def take_order():
  cost=0
  description=""
  pizza=input("Enter your pizza choice from menu: ").capitalize()
  if pizza=="Classic":
    ordered_pizza=Classic()
    cost=ordered_pizza.get_cost()
    description=ordered_pizza.get_description()
  elif pizza=="Margherita":
    ordered_pizza=Margherita()
    cost=ordered_pizza.get_cost()
    description=ordered_pizza.get_description()
  elif pizza=="Turkish pizza":
    ordered_pizza=Turk()
    cost=ordered_pizza.get_cost()
    description=ordered_pizza.get_description()
  elif pizza=="Plain":
    ordered_pizza=Plain()
    cost=ordered_pizza.get_cost()
    description=ordered_pizza.get_description()
  else:
    print("This order is unacceptable. Try again please")
  while True:
    topping=input("Enter your toppings if you want to add if you dont want to add you can enter space for skip this part: ").capitalize()

    if topping=="Olives":
      Topping=Olives()
      cost+=Topping.get_cost()
      description=description+" "+Topping.get_description()
    elif topping=="Mushrooms":
      Topping=Mushrooms()
      cost+=Topping.get_cost()
      description=description+" "+Topping.get_description()
    elif topping=="Goat cheese":
      Topping=Goat_Cheese()
      cost+=Topping.get_cost()
      description=description+" "+Topping.get_description()
    elif topping=="Meat":
      Topping=Meat()
      cost+=Topping.get_cost()
      description=description+" "+Topping.get_description()
    elif topping=="Onion":
      Topping=Onion()
      cost+=Topping.get_cost()
      description=description+" "+Topping.get_description()
    elif topping=="Corn":
      Topping=Corn()
      cost+=Topping.get_cost()
      description=description+" "+Topping.get_description()
    else:
      cost+=0
      break
  print("Cost:",cost,"TL","Description:",description)
  name=""
  credit_card_number=0
  credit_card_password=0
  id_number=0
  name=input("Please enter your name: ").capitalize()
  id_number=int(input("Please enter your id number: "))
  credit_card_number=int(input("Please enter your credit card number: "))
  credit_card_password=int(input("Please enter your card's password: "))
  order_time=datetime.datetime.now()
  infos=[name, description, cost, id_number, credit_card_number, credit_card_password, order_time]
  header=["Name","Descriptions","Cost","ID_Number","Credit_Card_Number","Credit_Card_Password","Time"]
  with open('Order_Database.csv','w') as file:
    saver=csv.writer(file)
    saver.writerow(header)
    saver.writerow(infos)

  
#This is our super class which called pizza and other subclasses inherited by this one
class Pizza(object):
  def __init__(self):
    self.description="Description"
    self.cost=0
  def get_cost(self):
    return self.cost
  def get_description(self):
    return self.description

class Classic(Pizza):
  def __init__(self):
    super().__init__()
    self.description="Normal pizza with cheese tomato sauce and toppings like olive pepper etc."
    self.cost=20

class Margherita(Pizza):
  def __init__(self):
    super().__init__()
    self.description="Pizza with the mozzarella cheese"
    self.cost=25


class Turk(Pizza):
  def __init__(self):
    super().__init__()
    self.description="This pizza includes Turkish Sucuk it's the difference from Classic Pizza"
    self.cost=30


class Plain(Pizza):
  def __init__(self):
    super().__init__()
    self.description="Normal pizza with cheese tomato sauce and toppings like olive pepper etc."
    self.cost=15
#This is our superclass which called Decorator and other subclasses inherited by this one
class Decorator(object):
  def __init__(self):
    self.description="Toppings"
    self.cost=0
  def get_cost(self):
    return self.cost
  def get_description(self):
    return self.description

class Olives(Decorator):
  def __init__(self):
    super().__init__()
    self.description
    self.cost=4

class Mushrooms(Decorator):
  def __init__(self):
    super().__init__()
    self.cost=3

class Goat_Cheese(Decorator):
  def __init__(self):
    super().__init__()
    self.cost=5

class Meat(Decorator):
  def __init__(self):
    super().__init__()
    self.cost=5

class Onion(Decorator):
  def __init__(self):
    super().__init__()
    self.cost=3

class Corn(Decorator):
  def __init__(self):
    super().__init__()
    self.cost=3

if __name__=="__main__":
  welcome_to_ordering_system()
  menu()
  take_order()
  see_you()
```
